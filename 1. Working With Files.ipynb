```python
# Deleting particular line in text files

file = open("test2.txt", 'w')

remove_text = input("Enter the text you want to remove")

for line in data:
  if remove_text in line:
    line = line.replace(remove_text, '')

  file.write(line)
file.close()
file = open("test2.txt", 'r')
file.seek(0)
file.read()

file.close()
```

## **JSON Read and Write**

```python
data = [
  {
    "name": "Alice Johnson",
    "email": "alice.johnson@example.com",
    "id": 101
  },
  {
    "name": "Brian Smith",
    "email": "brian.smith@example.com",
    "id": 102
  },
  {
    "name": "Catherine Lee",
    "email": "catherine.lee@example.com",
    "id": 103
  }
]
```

```python
import json

with open('json_write.json', 'w')as f:
  json.dump(data, f)
```

```python
with open('json_write.json', 'r')as f:
  json_data = json.load(f)
  print(json_data)
```
