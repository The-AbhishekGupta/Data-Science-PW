## **Exception Handling**

```python
class MyException(Exception):

  def __init__(self, msg, code=100):
    super().__init__(msg)
    self.code = code

def divide(a,b):
  if b==0:
    raise MyException("Not divisible my 0", code=120)
  else:
    return a//b

try:
  a, b = map(int, input("Enter two numer : ").split())
  print(divide(a, b))
except MyException as e:
  logging.debug("Trying to divide by 0")
```
