### **Matrix Multiplication**

```python
mat1 = np.array([[1,2,3,4], [5,6,7,8]])
mat2 = np.array([[4,5], [6,5], [5,9], [1,3]])
```

```python
# Method 1

mat3 = mat1 @ mat2
print(mat3)
```

```python
# Method 2

mat3  = np.matmul(mat1, mat2)
print(mat3)
```

```python
# Method 3
mat3 = np.dot(mat1,mat2)
print(mat3)
```

**Deep Copy vs Shallow Copy**

```python
array = [2,4,656,546,12,2,9,54]
array = np.array(array)
```

```python
# Shallow Copy

arr2 = array.view()
arr2
```

```python
arr2[5] = 909
print(arr2)
print(array)
```

```python
deep_arr = array.copy()
deep_arr
```

```python
deep_arr[3]= 0
print(deep_arr)
print(array)
```

**Array creation from function**

```python
array =  np.fromfunction(lambda i,j :  i==j, (3,3))
array
```

```python
text = '1,2,3,4,5,6,7,8'
array = np.fromstring(text, sep=',', dtype=int)
array
```

```python
array =  np.fromiter([1,2,3,4], dtype=float)
array
```

```python
arr= [1,2,34]
arr = np.array(arr)

copy1 = np.array(arr)
print(copy1)

copy2 = np.asarray(arr)
print(copy2)

print(copy1.__add__)
print(copy2.__add__)
print(arr.__add__)
```
