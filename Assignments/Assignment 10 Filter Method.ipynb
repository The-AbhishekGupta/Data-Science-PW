## **Filter Method**

```python
import seaborn as sns

df = sns.load_dataset('titanic')

df.head(3)
```

```python
# columns with missing values

missing_values = df.isna().sum()
missing_values[missing_values > 0]
```

```python
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])
df['embark_town'] = df['embark_town'].fillna(df['embark_town'].mode()[0])
```

```python
from sklearn.impute import KNNImputer

knn_imputer = KNNImputer(n_neighbors=3)
df['age_by_knnImputer'] =  knn_imputer.fit_transform(df[['age']])[:,0]
```

```python
df.head()
df['age'].mean(), df['age_by_knnImputer'].mean()
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

itr_impute = IterativeImputer(max_iter=10, random_state=42)
df['age_by_ItrImputer'] = itr_impute.fit_transform(df[['age', 'survived', 'pclass', 'sibsp', 'parch']])[:,0]
```

```python
df.head()
df['age'].mean(), df['age_by_ItrImputer'].mean()
```

```python
df.drop(['age', 'age_by_knnImputer', 'deck'], axis=1, inplace=True)

df.head(3)
```
