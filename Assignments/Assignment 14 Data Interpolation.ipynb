## **Data Interpolation**

Data interpolation is the process of estimating unknown values within a dataset based on the known values. In Python, there are various libraries available that can be used for data interpolation, such as NumPy, SciPy, and Pandas. Here is an example of how to perform data interpolation using the NumPy library:

<br>

***Linear Interpolation***

```python
import numpy as np
import seaborn as sns

x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])
```

```python
sns.scatterplot(x=x, y=y)
```

```python
# creating interploation

new_x = np.linspace(1,5,10)
y_interp = np.interp(new_x, x,y)

```

```python
sns.scatterplot(x=new_x, y=y_interp)
```

<br>

***Cubic Interpolation With Scipy***

```python
import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([1,8,27,64,125])

sns.scatterplot(x=x, y=y)
```

```python
from scipy.interpolate import interp1d

x_new = np.linspace(1,5,10)

inter_pl = interp1d(x,y,kind='cubic')
y_interp = inter_pl(x_new)

sns.scatterplot(x=x_new, y=y_interp)
```

<br>

**Polynomial Interpolation**

```python
import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([1,4,9,16,25])

sns.scatterplot(x=x, y=y)
```

```python
x_new = np.linspace(1,5,10)

p = np.polyfit(x, y, 4 )

y_new = np.polyval(p, x_new)
```

```python
sns.scatterplot(x=x_new, y=y_new)
```
