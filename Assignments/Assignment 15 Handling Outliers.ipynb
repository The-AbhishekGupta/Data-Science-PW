***We use Z-score to detect outliers***

````python
import numpy as np
from scipy import stats

z_score = np.abs(stats.zscore(marks))
z_score

threshold_z  = 2
````

````python
outliers_index = np.where(z_score > threshold_z)[0]
````

````python
marks = marks[:outliers_index[0]]
marks
````

***We use IQR Method to detect outliers***

````python
import numpy as np

marks=[45,32,56,75,89,54,32,89,90,87,67,54,45,98,99,67,74,1000,1100]
````

````python
import numpy as np

marks = [45,32,56,75,89,54,32,89,90,87,67,54,45,98,99,67,74,1000,1100]

Q1 = np.percentile(marks, 25)
Q3 = np.percentile(marks, 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

marks  = [x for x in marks if x >= lower_bound and x<= upper_bound]

marks
````
