````python
import pandas as pd

def convert_to_categorical(file_path, column_name, category_order):

  try:
    data = pd.read_csv(file_path)
    column_values = data[column_name]
    if column_values != set(category_order):
      category_order = input("RENTER : You category values mismatch with original values : ")

    pd.Categorical(df[column_name], categories=category_order, ordered=True)
    display(df.sort_values(by=column_name))

  except Exception as e:
    display(f'Error {e}')

file_path = input("Enter file path ")
column_name =  input('Enter column name')
category_order =  list(input("Enter category order").split())

convert_to_categorical(file_path, column_name, category_order)
````

````python
def convert_to_categorical_and_sort(file_path, column_name, category_order):
    try:
        df = pd.read_csv(file_path)
        df[column_name] = pd.Categorical(df[column_name], categories=category_order, ordered=True)
        sorted_df = df.sort_values(by=column_name)
        return sorted_df
    except Exception as e :
        print(f"Error: {e}")


if __name__ == "__main__":
    file_path = input("Enter the CSV file path: ")
    column_name = input("Enter the column name to convert to categorical: ")
    category_order = input("Enter the category order (comma-separated values): ")
    category_order = [cat.strip() for cat in category_order.split(",")]

    sorted_data = convert_to_categorical_and_sort(file_path, column_name, category_order)
    print("\nSorted Data:")
    print(sorted_data)
````


Q10. Write a Python program that reads a CSV file containing sales data for different products and visualizes the data using a stacked bar chart to show the sales of each product category over time. The program should prompt the user to enter the file path and display the chart.

````python
import pandas as pd
import matplotlib.pyplot as plt



# Step 3: Convert 'Date' to datetime format for better sorting/plotting
data['Date'] = pd.to_datetime(data['Date'], dayfirst=True)

# Step 4: Convert 'Units Sold' to numeric
data['Units Sold'] = pd.to_numeric(data['Units Sold'], errors='coerce')

# Step 5: Group data by Date and Product Category, summing Units Sold
grouped = data.groupby(['Date', 'Product Category'])['Units Sold'].sum().unstack(fill_value=0)

# Step 6: Plot a stacked bar chart
grouped.plot(kind='bar', stacked=True)

# Step 7: Customize the chart
plt.title('Sales of Each Product Category Over Time')
plt.xlabel('Date')
plt.ylabel('Units Sold')
plt.xticks(rotation=45)
plt.legend(title='Product Category')
plt.tight_layout()

# Step 8: Display the chart
plt.show()
````
