```python
import numpy as np
import matplotlib.pyplot as plt

# Create 100 points between 1 and 50
X = np.linspace(1, 50, 100)

# Calculate mean & standard deviation
mean = np.mean(X)
sd = np.std(X)

# Define the normal distribution function (using the correct formula)
def normal_dist(x, mean, sd):
    return (1 / (np.sqrt(2 * np.pi) * sd)) * np.exp(-0.5 * ((x - mean) / sd) ** 2)

# Compute probability density values
pdf = normal_dist(X, mean, sd)

# Plot the result
plt.figure(figsize=(6, 4))
plt.plot(X, pdf, color='red', linewidth=2)
plt.title("Normal Distribution", fontsize=14)
plt.xlabel("Values of Random Variable X")
plt.ylabel("Probability Density")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings


# Generate 1000 random numbers from a Uniform(0, 1) distribution
data = np.random.uniform(low=0, high=1, size=1000)

# Plot
plt.figure(figsize=(6,4))
sns.displot(data, kde=True, color='red')
plt.title("Continuous Uniform Distribution", fontsize=14)
plt.xlabel("Random Variable (X)")
plt.ylabel("Frequency")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
