Mutable objects can be modified after its created whereas immutable objects cannot be modified once created.

```python
'''
Mutable mean what gives us the ability of modify its value
Immutable mean what doesn't allow us to modify


list is Mutable
Tuple and String is Immutable

'''

checkMutable =  [1,2,3,4,5]
checkMutable[3] = 1999

print(checkMutable)


check_Immutable =  (6,9,32,2,21,3)   # Not allowed
check_Immutable[2] = 21212

print(check_Immutable)
```
